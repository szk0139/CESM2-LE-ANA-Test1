```python
import xarray as xr
import matplotlib.pyplot as plt

fn = "/glade/campaign/collections/gdex/data/d651056/CESM2-LE/atm/proc/tseries/month_1/TREFHT/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.cam.h0.TREFHT.185001-185912.nc"
ds = xr.open_dataset(fn)

# inspect
display(ds)
```

```python
# first monthly field
plt.figure(figsize=(8,4))
ds["TREFHT"].isel(time=0).plot()
plt.title(str(ds["time"].isel(time=0).values))
plt.show()
```

```python
# global/simple mean time series
plt.figure(figsize=(10,4))
ds["TREFHT"].mean(dim=["lat","lon"]).plot()
plt.title("Mean TREFHT time series")
plt.show()
```
